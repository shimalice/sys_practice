### Linear Regression

#### 独立変数と従属変数の値が共に知られている多数の標本（訓練データ）をもとに、独立変数から従属変数の値を求める（予測する）関数を学習するタスクを回帰と呼ぶ。線形回帰は回帰に対するもっともシンプルな方法のひとつであり、学習される関数が一次関数となるものである。一次関数であるため、データから求められる（学習される）のは各独立変数に対する重みと定数項（バイアス項）である。

```python
#!/usr/bin/env python
# linear regression to abalone data

from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np

# read data
datalist = []
f = open('pyml/abalone.data', 'r')
for line in f:
    line = line.rstrip()
    elems = line.split(',')
    datalist.append(elems[1:])
f.close()
data = np.array(datalist)
# length
x = np.c_[np.float_(data[:,0])]
# rings
y = np.c_[np.float_(data[:,-1])]

lr = linear_model.LinearRegression()
lr.fit(x, y)

# mesh for x_test
x_test = np.linspace(x.min(), x.max(), 1000)

#coef_: 傾き, intercept_: 切片
predicted = lr.coef_[0] * x_test + lr.intercept_ * np.ones(x_test.shape)

fig, ax = plt.subplots()
ax.scatter(x, y, c='red', marker='o', label='observed', lw=0)
ax.scatter(x_test, predicted, c='blue', marker='o', label='predicted', lw=0)

ax.tick_params(axis='both', which='major', labelsize=18)
ax.tick_params(axis='both', which='minor', labelsize=18)
ax.legend(loc='upper left', fontsize=18)
ax.set_xlabel('length', fontsize=18)
ax.set_ylabel('rings', fontsize=18)

plt.show()
```

#### 上記のプログラムにおいて、lr.fit(x, y)が実際に線形回帰を行っている箇所であり、それによって得られた係数がlr.coef_に、定数項（回帰直線のy 切片）がlr.intercept_に格納される。回帰直線を描くため、np.linspace(x.min(), x.max(), 1000)によってx の最小値から最大値までを1000 区間に等分割して得られる数列x_test を作っている。これとlr.coef_とlr.intercept_によって（x_test のそれぞれの値に対する）予測値を並べたベクトルpredicted を作成している。ax.scatter によって散布図（scatter plot）を描くことで無数の点が直線上に並び、回帰直線として見ることができる。

````python
#!/usr/bin/env python
# linear regression to abalone data

from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np

# read data
datalist = []
f = open('pyml/abalone.data', 'r')
for line in f:
    line = line.rstrip()
    elems = line.split(',')
    datalist.append(elems[1:])
f.close()
data = np.array(datalist)
# length
x = np.c_[np.float_(data[:,0])]
# rings
y = np.c_[np.float_(data[:,-1])]

x_train = x[0:800]
y_train = y[0:800]
x_test = x[800:1000]
y_test = y[800:1000]

lr = linear_model.LinearRegression()
lr.fit(x_train, y_train)

x_test_mat = x_test.reshape((x_test.shape[0], 1))
predicted = lr.predict(x_test_mat)
# mesh for x_test
#x_test = np.linspace(x.min(), x.max(), 1000)

#coef_: 傾き, intercept_: 切片
#predicted = lr.coef_[0] * x_test + lr.intercept_ * np.ones(x_test.shape)

fig, ax = plt.subplots()
ax.scatter(x, y, c='red', marker='o', label='observed', lw=0)
ax.scatter(x_test, predicted, c='blue', marker='o', label='predicted', lw=0)

ax.tick_params(axis='both', which='major', labelsize=18)
ax.tick_params(axis='both', which='minor', labelsize=18)
ax.legend(loc='upper left', fontsize=18)
ax.set_xlabel('length', fontsize=18)
ax.set_ylabel('rings', fontsize=18)

plt.show()
````

````python
cat pyml/gpAirQuality.py
````

#### 上記コードのax.fill では95%信頼区間を緑色に塗りつぶしている。fill では点の列を与えることでその内側が塗りつぶされる。[x_test, x_test[::-1]によってx 座標の値がx_testの最初の要素から順に辿っていき、最後の要素まで行った後、また最初の要素まで（[::-1]によって-1 ずつ動くので）戻っていく。y 座標に関しても同様に最初の要素から最後の要素まで行き、また戻ってくる。GP 回帰では独立変数の各値について従属変数の値がガウス分布に従うとモデル化される。回帰の結果、独立変数の各値について従属変数の期待値y_pred と分散sigmaが求まる。y_pred に±1.96 * sigma を足すことで95%信頼区間が描ける。ax.plot では予測値y_pred が青で、ax.scatter では訓練データy_train が赤で描かれている。

## 単純パーセプトロン

#### 単純パーセプトロンはもっとも単純なニューラルネットワークであり、小規模なデータセットに対する分類タスクに利用できる。以下のコードbinSimpPercep.py ではUCI MachineLearning Repository から得られるiris データセット(iris.data)とkeras を用い、アヤメの各部位の大きさからその品種を予測する単純パーセプトロンを訓練し、評価している。プログラムの実行時、同じディレクトリにiris.data を置いておく必要がある。iris データセットには3 つの品種のアヤメから得られたデータであるが、ここでは最初の2 つの品種だけを抜き出し、2 値分類を行っている。

````python
cat pyml/binSimpPercep.py
````

#### 上記のプログラムではle = preprocessing.LabelEncoder(), le.fit, le.transformによって文字列ラベル（アヤメの品種）を数値ラベル（0〜1）に変換している。さらにX =np.float_(np.array(datalist))ならびにy = np.array(depVar)において独立変数と従属変数をnumpy の配列に変換している。
#### model = Sequential()は層を重ねていく形のニューラルネットワークを作ることを決めている。model に密（dense）な中間層と活性化関数（activation function）を重ねることで単純パーセプトロンを実現している。model.compile において最適化手法はsgd（確率勾配降下法）、目的関数はbinary cross-entropy（二値交差エントロピー）、評価基準はaccuracy（精度）を使うように設定している。model.fit において実際の訓練と評価を行っている。
#### 学習済みのモデルを利用して予測を行うにはたとえば以下のようにする。
#### model.predict(X[5:6,:])
#### 引数として与える配列X を5:6 でスライスしているので、5 行目のみが切り出される。そのため上記関数の戻り値はX における5 番目のサンプルがカテゴリ1 に属する確信度を表している。出力にsigmoid 関数を使っているので、戻り値は0 から1 の範囲を取る。これが0.5 未満の場合はカテゴリ0 に属する可能性が高く、0.5 より大きい場合はカテゴリ1 に属する可能性が高いと考えられる。ここでは訓練データX を使っているが、実用上は独立変数の値は分かるが従属変数（正解ラベル）y が分からないデータに対して予測を行うことに意味がある。

````python
run pyml/binSimpPercep.py
````

## 多層パーセプトロン

#### 多層パーセプトロンはニューロンの層を重ねることでより複雑な分布を持つデータの分類を可能にしたものである。以下のコードmultiLayerPercep.py では中間層を2 つ重ねているため、多層パーセプトロンになっている。また、前述の単純パーセプトロンの例と異なり、アヤメの3つの品種（3 つのカテゴリ）への分類、すなわち多値分類を行っている。

````python
cat pyml/multiLayerPercep.py
````

#### 上記のプログラムではmodel.add(Dense... )という行が2 つあり、Dense（密）な層が2 つ重ねられているため、多層パーセプトロンである。多層パーセプトロンを使う場合、ニューロンを何層重ねるか、また中間層のノード数（上記のプログラムではnb_internalnodes）をいくつに定めると精度が出るかは対象とするタスクやデータの性質によって異なり、工夫が必要となる。試行錯誤によって決められることも多いが、自動で最適化する方法も研究されている。
#### 層を大量に重ねていったものが深層学習（ディープラーニング）である。
#### 多値分類のため、最終出力を行う活性化関数（activation）は（sigmoid の一般化である）softmax、損失関数としてcategorical cross-entropy を使うように設定している。また、前述の単純パーセプトロンの例では重み更新の回数（エポック数, nb_epoch）としてデフォルト値である 10 を使用したが、今回の例では精度を出すためにより多くの重み更新が必要のため、nb_epoch を100 に設定している。
#### ニューラルネットワークで多値分類を行う場合、従属変数を整数ではなくone-hot 表現（ひとつの成分のみが1 であり、残りの成分がすべて0 であるベクトル）として渡すことが多く、keras でもその形を取っている。そのための変換を行う関数がnp_utils.to_categoricalであり、これを使ってy をy_oneHot に変換している。たとえばカテゴリ数が3 でラベルy が2 の場合、そのone-hot 表現y_oneHot は(0,1,0)T になる。
#### 学習済みのモデルを利用して予測を行うにはたとえば以下のようにする。
#### model.predict(X[7:8,:])
#### 配列X を7:8 でスライスしているので、7 行目のみが切り出される。結果として得られる配列はX における7 番目のサンプルがそれぞれのカテゴリに属する確信度を表している。softmax 関数の出力であるので、すべてのカテゴリについて値を足すと1 になる。引数として与えたサンプルはもっとも値の大きいカテゴリに属する可能性が高いと考えられる。

````python
run pyml/multiLayerPercep.py
````

### MNIST データセットを使った文字認識

```python
'''Trains a simple convnet on the MNIST dataset.

Gets to 99.25% test accuracy after 12 epochs
(there is still a lot of margin for parameter tuning).
16 seconds per epoch on a GRID K520 GPU.
'''

from __future__ import print_function
import numpy as np
np.random.seed(1337)  # for reproducibility
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D
from keras.utils import np_utils
import pickle

batch_size = 128
nb_classes = 10
nb_epoch = 1

# input image dimensions
img_rows, img_cols = 28, 28
# number of convolutional filters to use
nb_filters = 32
# size of pooling area for max pooling
nb_pool = 2
# convolution kernel size
nb_conv = 3

# the data, shuffled and split between train and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')

# convert class vectors to binary class matrices
Y_train = np_utils.to_categorical(y_train, nb_classes)
Y_test = np_utils.to_categorical(y_test, nb_classes)

model = Sequential()

model.add(Convolution2D(nb_filters, nb_conv, nb_conv,
                        border_mode='valid',
                        input_shape=(1, img_rows, img_cols)))
model.add(Activation('relu'))
model.add(Convolution2D(nb_filters, nb_conv, nb_conv))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(nb_pool, nb_pool)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(nb_classes))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adadelta',
              metrics=['accuracy'])

model.fit(X_train, Y_train, batch_size=batch_size, nb_epoch=nb_epoch,
          verbose=1, validation_data=(X_test, Y_test))

# save the model and its weights
model.save_weights('model_weights_mnist_cnn.h5')
json_string = model.to_json()
with open('json_model_mnist_cnn.pkl','wb') as f:
  pickle.dump(json_string, f)
```
